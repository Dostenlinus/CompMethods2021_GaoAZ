# Simulation of a multi-agent sequential bargaining game

Here we demonstrate the Chen and Zapal model with N = 3 and q = 2. For simplicity, suppose $\delta$ = 0.2.
Assume y = 10, $x_1 = 1$, $x_3 = 10$, and consider a sequence of $x_2$, {1, 2, ..., 10}. We compare the transfer that each seller receives.

```python
import matplotlib.pyplot as plt

delta = 0.2
y = 10
x1 = 1
x3 = 10
x2s = []

for i in range(10):
    x2s.append(i+1)
```

### Transfer Promises

From the setup, we can observe that $x_2$ is indispensible at the beginning.


```python
# Transfer promises
t1_tran = []
t2_tran = []
t3_tran = []

for i in range(10):
    t1_tran.append((1-delta)*x1)
    
    if x2s[i] < (1-delta)*x3:
        t2_tran.append(x2s[i])
        t3_tran.append(0)
    else:
        t2_tran.append(0)
        t3_tran.append((1-delta)*x3)

fig1, ax = plt.subplots()
ax.plot(x2s, t1_tran, label = "Seller 1's transfer")
ax.plot(x2s, t2_tran, label = "Seller 2's transfer")
ax.plot(x2s, t3_tran, label = "Seller 3's transfer")
ax.set_xlabel("Seller 2's cost")
ax.set_ylabel("Transfer Promises")
ax.legend()
plt.show()

fig1.savefig("Figures/sampleTP.png")
```

### Up-front Payment

We can see that Seller 2 is more likely to be ignored in the Up-front Payment case since it is easier to exploit the seller with highest cost.

```python
# Up-front payment
t1_up = []
t2_up = []
t3_up = []

for i in range(10):
    t1_up.append(delta*delta*x1)
    
    if x2s[i] < delta*(1-delta)*x3:
        t2_up.append(x2s[i])
        t3_up.append(0)
    else:
        t2_up.append(0)
        t3_up.append(delta*(1-delta)*x3)

fig2, ax = plt.subplots()
ax.plot(x2s, t1_up, label = "Seller 1's payment")
ax.plot(x2s, t2_up, label = "Seller 2's payment")
ax.plot(x2s, t3_up, label = "Seller 3's payment")
ax.set_xlabel("Seller 2's cost")
ax.set_ylabel("Up-front Payments")
ax.legend()
plt.show()

fig2.savefig("Figures/sampleUP.png")
```
